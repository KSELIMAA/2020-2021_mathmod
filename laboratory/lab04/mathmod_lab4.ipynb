```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
g = 3.9
w = 2.9 # w^2
```

```python
def f(t):
    return 0.9*np.cos(2*t)
```

```python
def dx(x, t):
    return np.array([x[1], -w*x[0]-g*x[1]-f(t)])
```

```python
t = np.linspace(0, 29, 600)
x0 = np.array([0, -1.4])

x = odeint(dx, x0, t)
```

```python
plt.plot(x[:, 0], x[:, 1])
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
