```python
x0 = 111111
y0 = 99999
t0 = 0
a = 0.60
b = 0.77
c = 0.44 #1 случай
#c = 0.22
h = 0.66 #1 случай
#h = 0.88
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def P(t):
    p = np.sin(t+11) #1 случай
    #p = np.sin(22*t)
    return p
```

```python
def Q(t):
    q = np.cos(t+11) #1 случай
    #q = np.cos(22*t)
    return q
```

```python
def syst(y, t):
    return np.array([-a*y[0]-b*y[1]+P(t),-c*y[0]-h*y[1]+Q(t)]) #1 случай
    #return np.array([-a*y[0]-b*y[1]+P(t),-c*y[0]*y[1]-h*y[1]+Q(t)])
```

```python
t = np.linspace(0, 2)
v0 = np.array([x0, y0])

res = odeint(syst, v0, t)
```

```python
plt.plot(t, res[:, 0])
plt.plot(t, res[:, 1])
plt.xlabel("время")
plt.ylabel("численность армии")
plt.title("модель боевых действий №1")
plt.legend(["войска x", "войска y"])
plt.show()
```
